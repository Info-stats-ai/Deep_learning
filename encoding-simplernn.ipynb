```python
import numpy as np

docs = ['go india',
		'india india',
		'hip hip hurray',
		'jeetega bhai jeetega india jeetega',
		'bharat mata ki jai',
		'kohli kohli',
		'sachin sachin',
		'dhoni dhoni',
		'modi ji ki jai',
		'inquilab zindabad']
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer(oov_token='<nothing>')
# handle the words that is not there
```

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_index
# out of vocab is always assigned index 1
# tokenizer assigns the words according to the frequency
```

```python
tokenizer.word_counts
```

```python
tokenizer.document_count
```

```python
# converting words into sequence
sequences = tokenizer.texts_to_sequences(docs)
sequences
# these vectors are of varied length
 #index value is assigned according to the frequency
 # vector represnttion is in form mentioning there index at each word
```

```python
from keras.utils import pad_sequences
```

```python
sequences = pad_sequences(sequences,padding='post')
```

```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
```

```python
(X_train,y_train),(X_test,y_test) = imdb.load_data()
```

```python
print(X_train[0]) # data is preprocessed and integer encoded
```

```python
print(X_train.shape)
```

```python
print(len(X_train[2]), len(X_train[1]))
```

```python
# making the lengths equal
X_train = pad_sequences(X_train,padding='post',maxlen=50)
X_test = pad_sequences(X_test,padding='post',maxlen=50)
# review is trimmed to 50 length
# I will do it again
#The pad_sequences function reshaped X_train into a 2-dimensional array (25000, 50) by:

#Making each of the 25,000 sequences an inner dimension.
#Padding (adding zeros) or truncating each sequence so that they
#all have the same length of 50, as specified by maxlen=50.
```

```python
X_train[0]
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(SimpleRNN(32,input_shape=(50,1),return_sequences=False))
# there are 50 words in each review  so 50 time steps and 1 as single input
#feature
# return sequences = false, menas out generated at each time stamp will be provided
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

model.fit(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```
