```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense,LSTM,GRU, BatchNormalization, Bidirectional
```

```python
# Load the IMDb dataset
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=10000)

# Pad sequences to have the same length
x_train = pad_sequences(x_train, maxlen=100)
x_test = pad_sequences(x_test, maxlen=100)
```

```python
x_train.shape
```

```python
model = Sequential([
    Embedding(input_dim=10000, output_dim=32, input_shape=(100,)),  # Use input_shape instead of input_length
    Bidirectional(SimpleRNN(5)),
    Dense(1, activation='sigmoid')
])

model.summary()

```

```python
# Compile the LSTM model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
# Fit the LSTM model
history = model.fit(x_train, y_train, epochs=5, batch_size=32, validation_data=(x_test, y_test))
```

```python
# Define the LSTM model
model = Sequential([
    Embedding(10000, 32, input_shape=(100,)),
    LSTM(5, return_sequences=True),
    LSTM(5),
    Dense(1, activation='sigmoid')
])

model.summary()
```

```python


# Compile the LSTM model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
# Fit the LSTM model
history = model.fit(x_train, y_train, epochs=5, batch_size=32, validation_data=(x_test, y_test))
```

```python
# Define the GRU model
model = Sequential([
    Embedding(10000, 32, input_shape=(100,)),
    GRU(5, return_sequences=True),
    GRU(5),
    Dense(1, activation='sigmoid')
])

model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
history = model.fit(x_train, y_train, epochs=5, batch_size=32, validation_split=0.2)
```
