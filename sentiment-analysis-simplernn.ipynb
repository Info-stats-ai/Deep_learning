```python
# prompt: I want run this on streamlit

import streamlit as st
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, Embedding, Flatten
from tensorflow.keras.datasets import imdb
import numpy as np
import os # Import the os module for saving/loading

# Function to build and potentially load the model
@st.cache_resource
def get_model():
    model_path = 'sentiment_rnn_model.h5'
    model = Sequential()
    model.add(Embedding(10000, 2))
    model.add(SimpleRNN(32, return_sequences=False))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])

    # Check if the model exists and load it
    if os.path.exists(model_path):
        st.write("Loading pre-trained model...")
        model.load_weights(model_path)
    else:
        st.write("Training model (this may take a few minutes)...")
        # Load and preprocess data for training
        (X_train, y_train), (X_test, y_test) = imdb.load_data()
        X_train = pad_sequences(X_train, padding='post', maxlen=50)
        X_test = pad_sequences(X_test, padding='post', maxlen=50)
        # Train the model
        model.fit(X_train, y_train, epochs=5, validation_data=(X_test, y_test))
        # Save the trained model
        model.save_weights(model_path)
        st.write("Model trained and saved.")

    return model

# Load the IMDB word index
@st.cache_resource
def get_imdb_word_index():
  return imdb.get_word_index()

imdb_word_index = get_imdb_word_index()

# Function to encode text using the IMDB word index
def encode_text(text, word_index, maxlen=50):
  words = text.lower().split()
  # Handle potential KeyError by using .get and providing a default for unknown words (index 2 is unknown)
  encoded = [word_index.get(word, 2) + 3 for word in words] # +3 because 0, 1, 2 are reserved in IMDB data
  padded = pad_sequences([encoded], padding='post', maxlen=maxlen)
  return padded[0] # return the single sequence


# Get the model
model = get_model()

# Streamlit app
st.title("Sentiment Analysis with SimpleRNN")

st.write("Enter a sentence to predict its sentiment (Positive or Negative).")

user_input = st.text_area("Enter your sentence here:", "")

if st.button("Analyze Sentiment"):
    if user_input:
        # Encode the user input
        encoded_input = encode_text(user_input, imdb_word_index, maxlen=50)
        # Reshape for the model input (batch size of 1)
        encoded_input = np.expand_dims(encoded_input, axis=0)

        # Predict sentiment
        prediction = model.predict(encoded_input)

        # Determine sentiment based on the sigmoid output
        if prediction[0] > 0.5:
            sentiment = "Positive"
            st.success(f"Predicted Sentiment: **{sentiment}** (Confidence: {prediction[0][0]:.4f})")
        else:
            sentiment = "Negative"
            st.error(f"Predicted Sentiment: **{sentiment}** (Confidence: {prediction[0][0]:.4f})")
    else:
        st.warning("Please enter some text to analyze.")

# Instructions on how to run
st.markdown("""
To run this Streamlit app in Google Colab:

1.  Save the code above as a Python file (e.g., `app.py`).
2.  Install Streamlit and related libraries:
```
